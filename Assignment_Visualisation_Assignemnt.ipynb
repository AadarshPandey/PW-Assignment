## 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(1)
categories = np.random.choice(['A', 'B', 'C', 'D'], 200)

values = np.zeros(200)
values[categories == 'A'] = np.random.normal(3, 1, sum(categories == 'A'))
values[categories == 'B'] = np.random.normal(5, 1, sum(categories == 'B'))
values[categories == 'C'] = np.random.normal(7, 1, sum(categories == 'C'))
values[categories == 'D'] = np.random.normal(9, 1, sum(categories == 'D'))

df = pd.DataFrame({'Categories': categories, 'Values': values})

sns.boxplot(x='Categories', y='Values', data=df)
plt.title('Box Plot of Values Across Categories')
plt.show()
```

## 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(2)
data = np.random.multivariate_normal([0, 0, 0], [[1, 0.8, 0.6], [0.8, 1, 0.5], [0.6, 0.5, 1]], 100)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

## 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.randint(10, 100, 50)
colors = np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], 50)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_label="Random Points")

show(p)
```

## 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears' ]

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9, color=factor_cmap('x', palette=Spectral6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

## 5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource
import numpy as np

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    z=data_heatmap.flatten()
))

p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y')

mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('z', 'Viridis256', data_heatmap.min(), data_heatmap.max()))

color_bar = ColorBar(color_mapper=mapper, label_standoff=12, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
